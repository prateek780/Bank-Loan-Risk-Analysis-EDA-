```python

target_1_df = df.query('TARGET == 1')
target_0_df = df.query('TARGET == 0')
```

```python
def histplot(col_name="AMT_CREDIT", num_ticks=50):
    plt.figure(figsize=[15, 7])
    
    plt.subplot(1, 3, 1)
    sns.histplot(df[col_name], bins=num_ticks, kde=True, color="skyblue")
    plt.xlabel(col_name)
    plt.ylabel("Frequency")
    
    # Comparison Plot
    plt.subplot(1, 3, 2)
    sns.histplot(target_0_df[col_name], bins=num_ticks, kde=True, color="green", label='TARGET = 0')
    sns.histplot(target_1_df[col_name], bins=num_ticks, kde=True, color="black", label='TARGET = 1')
    
    # Setting the axis.
    plt.xlabel(col_name)
    plt.ylabel("Frequency")
    plt.title(f"Univariate Analysis of {col_name} with TARGET=0 / TARGET=1")
    plt.legend()

    plt.tight_layout()
    plt.show()
```
